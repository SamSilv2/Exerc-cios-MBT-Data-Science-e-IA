# Exercícios

Exercício 1:
Listar todos os clientes e seus respectivos pedidos.

```python
%%sql

SELECT c.CustomerID, c.CompanyName, o.OrderID
FROM Customers AS c
LEFT JOIN Orders AS o
ON c.CustomerID = O.CustomerID
```

Exercício 2:
Listar todos os pedidos feitos por um cliente específico.

```python
%%sql

SELECT OrderID, OrderDate, ShippedDate
FROM Orders
WHERE CustomerID = 'ALFKI'
```

Exercício 3:
Listar todos os produtos e seus respectivos fornecedores.

```python
%%sql

SELECT p.ProductID, p.ProductName, p.SupplierID, s.CompanyName
FROM Products AS p
LEFT JOIN Suppliers AS s
ON p.SupplierID = s.SupplierID
```

Exercício 4:
Listar todos os produtos de uma determinada categoria.

```python
%%sql

SELECT p.ProductName, c.CategoryName
FROM Products AS p
INNER JOIN Categories AS c
USING(CategoryID)
WHERE CategoryName = 'Beverages'
```

Exercício 5:
Listar todos os ped
idos entregues por determinada transportadora (explore a tabela para escolher uma transportadora)

```python
%%sql

SELECT o.OrderID, o.OrderDate, s.CompanyName
FROM Orders AS o
INNER JOIN Shippers AS s
ON o.ShipVia = s.ShipperID
WHERE s.CompanyName = 'Speedy Express' AND o.ShippedDate IS NOT NULL
```

Exercício 6: Listar todos os clientes que fizeram um pedido após uma determinada data (explore a tabela para descobrir os períodos e defina uma data)

```python
%%sql

SELECT c.CustomerID, c.CompanyName, o.OrderID, o.OrderDate
FROM Customers AS c
LEFT JOIN Orders AS o
USING(CustomerID)
WHERE OrderDate > 2014-07-04
```

Exercício 7:
Listar todos os produtos que estão em estoque e que precisam ser reabastecidos

```python
%%sql

SELECT ProductName, UnitsInStock, ReorderLevel
FROM Products
WHERE UnitsInStock < ReorderLevel
```

Exercício 8: Listar todos os funcionários e seus respectivos supervisores.

```python
%%sql

SELECT e.EmployeeID, e.LastName, e.FirstName, s.LastName AS SupervisorLastName, s.FirstName AS SupervisorFirstName
FROM Employees AS e
LEFT JOIN Employees AS s
ON e.ReportsTo = s.EmployeeID
```

Exercício 9: Listar todos os fornecedores e seus respectivos produtos.

```python
%%sql

SELECT s.SupplierID, s.CompanyName, p.ProductID, p.ProductName
FROM Suppliers AS s
LEFT JOIN Products AS p
USING(SupplierID)
```

Exercício 10:
Listar o total de vendas por cliente.


```python
%%sql

SELECT c.CustomerID, c.CompanyName, SUM((od.Quantity*od.UnitPrice))*(1-od.Discount) AS TotalSales
FROM Customers AS c
LEFT JOIN Orders AS o
ON c.CustomerID = o.CustomerID
LEFT JOIN OrderDetails AS od
ON o.OrderID = od.OrderID
GROUP BY c.CustomerID
```
