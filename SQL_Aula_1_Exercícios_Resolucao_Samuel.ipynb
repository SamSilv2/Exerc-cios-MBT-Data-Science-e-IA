#Exercícios

Exercício 1:
Selecione todos os clientes da tabela "Customers".

```python
%%sql

SELECT * FROM Customers
```

Exercício 2:
Selecione o nome e o país de todos os clientes da tabela "Customers".

```python
%%sql

SELECT ContactName, Country
FROM Customers
```

Exercício 3:
Selecione todos os produtos da tabela "Products" que tenham preço maior que $10

```python
%%sql

SELECT * FROM Products
WHERE UnitPrice > 10
```

Exercício 4:
Selecione todos os funcionários da tabela "Employees" que tenham o título "Vice President, Sales".

```python
%%sql

SELECT * FROM Employees
WHERE Title = "Vice President, Sales"
```

Exercício 5:
Selecione todos os fornecedores da tabela "Suppliers" que não sejam do país "USA".

```python
%%sql

SELECT * FROM Suppliers
WHERE Country <> "USA"
```

Exercício 6:
Selecione todos os pedidos da tabela "Orders" que foram feitos no ano de 2014.

```python
%%sql

SELECT * FROM Orders
WHERE strftime('%Y', OrderDate) = '2014';
```

Exercício 7:
Selecione todos os produtos da tabela "Products" cujo nome comece com a letra 'A'.

```python
%%sql

SELECT * FROM Products
WHERE ProductName LIKE 'A%'
```

Exercício 8:
Selecione todos os funcionários da tabela "Employees" que nasceram antes de 1970.

```python
%%sql
SELECT * FROM Employees
WHERE DATE(BirthDate) < DATE(1970,1,1)
```

Exercício 9:
Selecione todos os clientes da tabela "Customers" que estão localizados em cidades que começam com a letra 'S'.

```python
%%sql

SELECT * FROM Customers
WHERE City LIKE 'S%'
```

Exercício 10:
Selecione todos os produtos da tabela "Products" cujo preço está entre $50 e $100.

```python
%%sql

SELECT * FROM Products
WHERE UnitPrice BETWEEN 50 AND 100
```

Exercício 11:
Selecione todas as categorias distintas da tabela "Categories".

```python
%%sql

SELECT DISTINCT CategoryName FROM Categories
```

Exercício 12:
Selecione os 10 primeiros clientes da tabela "Customers", ordenados por nome em ordem alfabética.

```python
%%sql
SELECT * FROM Customers
ORDER BY ContactName
LIMIT 10
```

Exercício 13: Selecione todos os pedidos da tabela "Orders" que foram enviados através do transportadora com o ID 3 e cujo frete seja menor que $10

```python
%%sql
SELECT * FROM Orders
WHERE ShipVia = 3 AND Freight <10
```

Exercício 14:
Selecione todos os fornecedores da tabela "Suppliers" que têm o nome começando com a letra 'M' ou terminando com a letra 'A'.

```python
%%sql

SELECT * FROM Suppliers
WHERE CompanyName LIKE 'M%' OR CompanyName LIKE '%A'
```

Exercício 15:
Selecione todos os produtos da tabela "Products" cujo ID de fornecedor está na lista de IDs (1, 3, 5).

```python
%%sql

SELECT * FROM Products
WHERE ProductID IN (1, 3, 5)
```
