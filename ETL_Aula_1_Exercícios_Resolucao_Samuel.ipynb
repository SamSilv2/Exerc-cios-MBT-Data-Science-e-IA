# Exercício

Faça a extração de alguma tabela da Wikipedia utilizando a biblioteca BeautifulSoup

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import sqlite3
import numpy as np
```

```python
url = "https://www.investing.com/equities/intel-corp-balance-sheet"
response = requests.get(url)
html = response.content

soup = BeautifulSoup(html, "html.parser")
soup
```

```python
table1 = soup.find_all("table", class_="genTbl reportTbl")[1]
table2 = soup.find_all("table", class_="genTbl reportTbl")[2]
table3 = soup.find_all("table", class_="genTbl reportTbl")[3]
table4 = soup.find_all("table", class_="genTbl reportTbl")[4]
table5 = soup.find_all("table", class_="genTbl reportTbl")[5]
```

```python
data1 = []
for row in table1.find_all("tr"):
    cols = row.find_all("td")
    if cols:
        values = [col.text.strip() for col in cols]
        data1.append(values)
data2 = []
for row in table2.find_all("tr"):
    cols = row.find_all("td")
    if cols:
        values = [col.text.strip() for col in cols]
        data2.append(values)
data3 = []
for row in table3.find_all("tr"):
    cols = row.find_all("td")
    if cols:
        values = [col.text.strip() for col in cols]
        data3.append(values)
data4 = []
for row in table4.find_all("tr"):
    cols = row.find_all("td")
    if cols:
        values = [col.text.strip() for col in cols]
        data4.append(values)
data5 = []
for row in table5.find_all("tr"):
    cols = row.find_all("td")
    if cols:
        values = [col.text.strip() for col in cols]
        data5.append(values)
```

```python
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
df3 = pd.DataFrame(data2)
df4 = pd.DataFrame(data3)
df5 = pd.DataFrame(data4)
df_intc_bs = pd.concat([df1, df2, df3, df4, df5])
df_intc_bs
```

```python
df_intc_bs.columns = ["Variable", "2023", "2022", "2021", "2020"]
df_intc_bs.set_index('Variable', inplace=True)
df_intc_bs
```

```python
df_intc_bs['2023'] = df_intc_bs['2023'].replace('-', np.nan)
df_intc_bs['2022'] = df_intc_bs['2022'].replace('-', np.nan)
df_intc_bs['2021'] = df_intc_bs['2021'].replace('-', np.nan)
df_intc_bs['2020'] = df_intc_bs['2020'].replace('-', np.nan)
df_intc_bs['2023'] = df_intc_bs['2023'].astype(float)
df_intc_bs['2022'] = df_intc_bs['2022'].astype(float)
df_intc_bs['2021'] = df_intc_bs['2021'].astype(float)
df_intc_bs['2020'] = df_intc_bs['2020'].astype(float)

df_intc_bs
```
